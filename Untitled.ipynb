```python
type = df['Type']
arr_type = type.values
id = df['ID']
arr_id = id.values
skip_id = []
arr_score = []
prev_score = 0
arr_flag = []
flag = 0
counter = 0
row = 0
```

```python
len(arr_id)
```

```python
for n in arr_type:
    if arr_id[row] in skip_id:
        pass
    else:
        if prev_score == 8 and n == 1:
            counter += 1
            flag = 1
            skip_id.append(arr_id[row])
        else:
            flag = 0
            prev_score += n
    arr_flag.append(flag)
    arr_score.append(prev_score)
    row += 1
```
